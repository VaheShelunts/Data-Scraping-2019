# Task 2

```python
wikipedia.set_lang("ru")
```

```python
Adele=wikipedia.summary("Adele")
Queen=wikipedia.summary("Queen")
Florence=wikipedia.summary("Florence")
```

```python
sw=stopwords.words("russian")
```

```python
Adele_blob=TextBlob(Adele)
Adele_words=[i for i in Adele_blob.words.lower() if i not in sw and i.isalnum()]
Adele_words=pd.Series(Adele_words)
Adele_words.value_counts()[:5] #most common words for Adele summary
```

```python
Queen_blob=TextBlob(Queen)
Queen_words=[i for i in Queen_blob.words.lower() if i not in sw and i.isalnum()]
Queen_words=pd.Series(Queen_words)
Queen_words.value_counts()[:5] #most common words for Queen summary
```

```python
Florence_blob=TextBlob(Florence)
Florence_words=[i for i in Florence_blob.words.lower() if i not in sw and i.isalnum()]
Florence_words=pd.Series(Florence_words).value_counts()[:5]
Florence_words #most common words for Florence summary
```
